```python
 %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib import pyplot as mp

# File to Load (Remember to change these)
city_csv = "city_data.csv"
ride_csv = "ride_data.csv"

city_data_df = pd.read_csv(city_csv)
ride_data_df = pd.read_csv(ride_csv)

#check data
city_data_df.head()
```

```python
ride_data_df.head()
```

```python
# Combine the data into a single dataset

city_ride_merge = pd.merge(ride_data_df, city_data_df)
city_ride_merge.head()
```

```python
# Average Fare ($) Per City
# Total Number of Rides Per City
# Total Number of Drivers Per City
# City Type (Urban, Suburban, Rural)

city_ride_gb = city_ride_merge.groupby(['type'])

#urban city averages 
urban_df = city_ride_gb.get_group('Urban')
urban_driver_avg = urban_df.groupby('city').agg({'fare' : 'mean', 'ride_id' : 'count'}).sort_values('fare', ascending=False)

#suburban 
suburban_df = city_ride_gb.get_group('Suburban')
suburban_driver_avg = suburban_df.groupby('city').agg({'fare' : 'mean', 'ride_id' : 'count'}).sort_values('fare', ascending=False)

#rural
rural_df = city_ride_gb.get_group('Rural')
rural_driver_avg = rural_df.groupby('city').agg({'fare' : 'mean', 'ride_id' : 'count'}).sort_values('fare', ascending=False)

urban_df.head(20)
```

```python
# build a Bubble Plot that showcases the relationship between four key variables
# You must stick to the Pyber color scheme (Gold, Light Sky Blue, and Light Coral) in producing your plot and pie charts.
# When making your Bubble Plot, experiment with effects like alpha, edgecolor, and linewidths.

s = [30*2**n for n in range(6)]

ax = urban_driver_avg.plot.scatter(x = 'ride_id',
                                   y = 'fare',
                                   c = 'lightcoral',
                                   edgecolors = 'black',
                                   s = s,
                                   label = 'Urban')

suburban_driver_avg.plot.scatter(x = 'ride_id',
                                 y = 'fare',
                                 c = 'gold',
                                 edgecolors = 'black',
                                 s = s,
                                 ax = ax,
                                 label = 'Suburban')

rural_driver_avg.plot.scatter(x = 'ride_id',
                              y = 'fare',
                              c = 'lightskyblue',
                              edgecolors = 'black',
                              s = s,
                              ax = ax,
                              label = 'Rural')

# Labels
plt.title('2018 Pyber Ride Sharing Data')
plt.xlabel('Total # of Rides Per City')
plt.ylabel('Average Fare ($)')
plt.grid()

# Ledger
plt.legend(title = 'City Types')


mp.savefig('pyber_data_2018.png')

# Show plot
plt.show()
```

```python
# In addition, you will be expected to produce the following three pie charts:


# % of Total Fares by City Type
# % of Total Rides by City Type
# % of Total Drivers by City Type
```

```python
# % of Total Fares by City Type

# Calculate percentages for TR
urban_fare_sum = urban_df['fare'].sum()
suburban_fare_sum = suburban_df['fare'].sum()
rural_fare_sum = rural_df['fare'].sum()

# TF Pie Chart
sizes = [urban_fare_sum, rural_fare_sum,suburban_fare_sum]
labels = 'Urban', 'Rural', 'Suburban'
colors = ['lightcoral', 'gold', 'lightskyblue']
explode = (0.1, 0.0, 0.0) 

# TF Pie Chart Settings
fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode = explode, 
        labels = labels, 
        colors = colors,
        autopct = '%1.2f%%',
        shadow = True, 
        startangle = 250)
plt.title('% of Total Fares by City Type')


mp.savefig('pie_chart_total_fares.png')
plt.show()
```

```python
# % of Total Rides by City Type

# Calculate Percents for TR
total_rides_list = [x[0] for x in [urban_df.shape, rural_df.shape, suburban_df.shape]]

# TF Pie Chart
labels = 'Urban', 'Rural', 'Suburban'
sizes = total_rides_list
colors = ['lightcoral', 'gold', 'lightskyblue']
explode = (0.1, 0.0, 0.0) 

# TR Pie Chart Settings
fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode = explode, 
        labels = labels, 
        colors = colors,
        autopct = '%1.2f%%',
        shadow = True, 
        startangle = 250)
plt.title('% of Total Rides by City Type')


mp.savefig('pie_chart_total_rides.png')
plt.show()
```

```python
# % of Total Drivers by City Type

# Calculate Percents for TD
urban_df_gb = urban_df.groupby('city')['driver_count'].unique().to_frame()
urban_df_driver_sum = urban_df_gb.sum()
suburban_df_gb = suburban_df.groupby('city')['driver_count'].unique().to_frame()
suburban_df_driver_sum = suburban_df_gb.sum()
rural_df_gb = rural_df.groupby('city')['driver_count'].unique().to_frame()
rural_df_driver_sum = rural_df_gb.sum()

# Pie Chart for TD
labels = 'Urban', 'Rural', 'Suburban'
sizes = [urban_df_driver_sum, rural_df_driver_sum,suburban_df_driver_sum]
colors = ['lightcoral', 'gold', 'lightskyblue']
explode = (0.1, 0.0, 0.0) 

# TD plot settings
fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode = explode, 
        labels = labels, 
        colors = colors,
        autopct = '%1.2f%%',
        shadow = True, 
        startangle = 220)
plt.title('% of Total drivers by City Type')


mp.savefig('pie_chart_total_drivers.png')
plt.show()
```
